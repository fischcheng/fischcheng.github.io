## 1.5 layer reduced-gravity model on equatorial beta-plane
A numerical 1.5 layer shallow water model to show thermocline perturbation evolution, as well as Kelvin and Rossby wave propagation and reflection. Domain was set to [30S-30N], [120E-80W], one-hour timestep, integration over 200days.

```python
import os
import matplotlib.pyplot as plt
import numpy as np
from netCDF4 import Dataset
from scipy.interpolate import griddata
from mpl_toolkits.basemap import Basemap
from mpl_toolkits.basemap import interp
from matplotlib import animation
```

```python
# Domaine and grid coordinates;
# Longitude
lon_w, lon_e, dlon, nlon = 120., 280., 1., 161
lons=arange(lon_w,lon_e+1,dlon)

# Latitude
lat_s, lat_n, dlat, nlat = -30., 30., 0.5, 121 ;
lats = arange(lat_s,lat_n+0.5,dlat)
mLON, mLAT = np.meshgrid(lons, lats)

# Spatial Smooth
xn=arange(lon_w,lon_e+0.25,0.25)
yn = arange(lat_s,lat_n+0.25,0.25)
ilon,ilat=np.meshgrid(xn,yn)
shape(ilon)
```

```python
# Parameters
r0=6.37e6 # Earth radius
D1=150.  # Original thermocline depth
gprime=3.7e-2   #reduced gravity

# Dissipation terms
alpha =1./(100.*86400.)
gamma =1./(100.*86400.)

# Degree to distance conversion
dy = dlat*np.pi/180*r0
dx = dlon*np.pi/180*r0
beta = (2.*7.3*1e-5*np.cos(0.))/r0

# Timestep
dt, days, ntsp =3600, 200*86400, 200*86400/3600  #dt one hour, integration 200days
ntspm = int(ntsp/24)+1
```

```python
# Mesh for coriolis term beta*u and beta*v on u and v grids
#y=fLAT*np.pi/180*r0
iy=arange(-30.25,30.5,0.5)
ix=arange(119.5,281,1)
fvLON, fvLAT = np.meshgrid(lons, iy)
fuLON, fuLAT = np.meshgrid(ix, lats)
beta_fu=beta*fuLAT*np.pi/180*r0  # for v calculation, actuall on u grid
beta_fv=beta*fvLAT*np.pi/180*r0  # on v grid
```

```python
# Construct u,v,h matrices
u = np.zeros((nlat, nlon+1, ntsp))   # u,v on the edges 
v = np.zeros((nlat+1, nlon, ntsp))  
h = np.zeros((nlat, nlon, ntsp))     # h in the grid box center
uu = np.zeros((nlat, nlon+1, ntspm))  # smoothed along time-dimension
vv = np.zeros((nlat+1, nlon, ntspm))
#hh = np.zeros((nlat, nlon, ntspm))
hh = np.zeros((241, 641, ntspm))
hs=np.zeros((241, 641,ntsp))


# B.C. for u, v of all time
v[0,:,:], v[nlat,:,:],u[:,0,:],u[:,nlon,:]  = 0.,0.,0.,0.
h[:,:,0] = 2*np.exp(-np.power((mLAT/2.5), 2)-np.power(((mLON-180)/10.), 2))
```

```python
# Integration scheme first timestep
u[0:nlat,1:nlon,1]=(1-dt*gamma)*u[0:nlat,1:nlon,0]+dt*\
(beta_fv[0:nlat,1:nlon]*(v[0:nlat,0:nlon-1,0]+v[0:nlat,1:nlon,0])+\
 beta_fv[1:nlat+1,1:nlon]*(v[1:nlat+1,0:nlon-1,0]+v[1:nlat+1,1:nlon,0]))/4.\
-gprime*(h[0:nlat,1:nlon,0]-h[0:nlat,0:nlon-1,0])/dx*dt

v[1:nlat,0:nlon,1]=(1-dt*gamma)*v[1:nlat,0:nlon,0]-dt*\
(((beta_fu[0:nlat-1,0:nlon])*(u[0:nlat-1,0:nlon,0]+u[0:nlat-1,1:nlon+1,0]))+\
((beta_fu[1:nlat,0:nlon])*(u[1:nlat,0:nlon,0]+u[1:nlat,1:nlon+1,0])))/4.\
-gprime*(h[1:nlat,0:nlon,0]-h[0:nlat-1,0:nlon,0])/dy*dt

h[0:nlat,0:nlon,1]=(1-dt*alpha)*h[0:nlat,0:nlon,0]-dt*D1*((u[0:nlat,1:nlon+1,0]-u[0:nlat,0:nlon,0])/dx+\
                                                           (v[1:nlat+1,0:nlon,0]-v[0:nlat,0:nlon,0])/dy)
hs[:,:,0]=interp(h[:,:,0], lons,lats, ilon,ilat, order=3)
hs[:,:,1]=interp(h[:,:,1], lons,lats, ilon,ilat, order=3)
hh[:,:,0]=hs[:,:,0]
```

```python
for it in range(2,2400):
    u[0:nlat,1:nlon,it]=(1-2*dt*gamma)*u[0:nlat,1:nlon,it-2]+2*dt*\
    (beta_fv[0:nlat,1:nlon]*(v[0:nlat,0:nlon-1,it-1]+v[0:nlat,1:nlon,it-1])+\
     beta_fv[1:nlat+1,1:nlon]*(v[1:nlat+1,0:nlon-1,it-1]+v[1:nlat+1,1:nlon,it-1]))/4.\
    -gprime*(h[0:nlat,1:nlon,it-1]-h[0:nlat,0:nlon-1,it-1])/dx*2*dt
    
    v[1:nlat,0:nlon,it]=(1-2*dt*gamma)*v[1:nlat,0:nlon,it-2]-2*dt*\
    (((beta_fu[0:nlat-1,0:nlon])*(u[0:nlat-1,0:nlon,it-1]+u[0:nlat-1,1:nlon+1,it-1]))+\
     ((beta_fu[1:nlat,0:nlon])*(u[1:nlat,0:nlon,it-1]+u[1:nlat,1:nlon+1,it-1])))/4.\
    -gprime*(h[1:nlat,0:nlon,it-1]-h[0:nlat-1,0:nlon,it-1])/dy*2*dt
    
    h[0:nlat,0:nlon,it]=(1-2*dt*alpha)*h[0:nlat,0:nlon,it-2]-dt*D1*((u[0:nlat,1:nlon+1,it-1]-u[0:nlat,0:nlon,it-1])/dx\
                                                              +(v[1:nlat+1,0:nlon,it-1]-v[0:nlat,0:nlon,it-1])/dy)
    # Spatial smoothing every timestep
    hs[:,:,it]=interp(h[:,:,it],lons,lats, ilon,ilat,order=1)

    if ((it)%24)==0:
        tn = (it+1)/24
        hh[:,:,tn]=(hs[:,:,it-2]+hs[:,:,it-1]+hs[:,:,it])/3.
        uu[:,:,tn]=(u[:,:,it-2]+u[:,:,it-1]+u[:,:,it])/3.
        vv[:,:,tn]=(v[:,:,it-2]+v[:,:,it-1]+v[:,:,it])/3.
    if it==4799:
        print("finished")
```
